# Headline

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

### Inspect the Data

```python
df = pd.read_csv("codon_usage.csv", low_memory=False)
df.head()
```

### Count occurences of distinct values in column 'Kingdom'

```python
df.groupby(by='Kingdom', as_index=True).agg({'Kingdom': pd.Series.count})
```

### Select column 'Kingdom'

```python
df_kingdom = df[["Kingdom"]]
df_kingdom.head()
```

### One Hot Encoding of column 'Kingdom'

```python
df_ohe = pd.get_dummies(df_kingdom)
df_ohe
```

### Apply the DGIM algorithm to the 'Kingdom_pri' column

```python
from dgim import Dgim
colname = "Kingdom_pri"
window_size = 1000

dgim = Dgim(N=window_size, error_rate=0.1)

for i, row in df_ohe.iterrows():
    if (row[colname]):
        dgim.update(True)
    else:
        dgim.update(False)

dgim_result = dgim.get_count()
dgim_result
```

### Calculate real number of ones

```python
df_ohe.iloc[-window_size:][colname].sum()
```
